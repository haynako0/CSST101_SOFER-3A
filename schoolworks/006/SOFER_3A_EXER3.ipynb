# Part 4: Visualize a Probability Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
import requests
from PIL import Image
from io import BytesIO

n = 5
p = 0.5

k = np.arange(0, n + 1)

probabilities = binom.pmf(k, n, p)

for i in range(len(k)):
    print(f"P(X = {k[i]}) = {probabilities[i]:.4f}")

image_url = "https://i.postimg.cc/G2dXDd96/wp12936728.jpg"
response = requests.get(image_url)
main_image = Image.open(BytesIO(response.content))
main_image = main_image.transpose(Image.FLIP_LEFT_RIGHT)

sticker_url = "https://i.postimg.cc/DwQ3JyvS/kuromi.png"
response = requests.get(sticker_url)
sticker_image = Image.open(BytesIO(response.content))

plt.figure(facecolor='#f8f8ff', figsize=(10, 6))

plt.bar(k, probabilities, color='#f4b5eb', edgecolor='black')

xlim = plt.gca().get_xlim()
ylim = plt.gca().get_ylim()

plt.imshow(main_image, extent=(xlim[0], xlim[1], ylim[0], ylim[1]), alpha=0.8, aspect='auto')

sticker = sticker_image.resize((50, 50))
plt.figimage(sticker, xo=0, yo=0, zorder=10)

plt.title('Probability Distribution of Getting 0-5 Heads in 5 Coin Flips', color='black', fontsize=16, fontweight='bold')
plt.xlabel('Number of Heads', color='black', fontsize=12)
plt.ylabel('Probability', color='black', fontsize=12)

plt.grid(False)

plt.show()
```
